```python
# Count orders by hour
hour_frequency = df_ords['order_hour_of_day'].value_counts().sort_index()
print("Orders by hour:")
print(hour_frequency)
```

```python
# Find the hour with maximum orders
busiest_hour = hour_frequency.idxmax()
max_orders = hour_frequency.max()

print(f"\nBusiest hour: {busiest_hour}:00")
print(f"Number of orders: {max_orders}")
```

```python
import matplotlib.pyplot as plt

# Create a bar plot
plt.figure(figsize=(12, 6))
hour_frequency.plot(kind='bar')
plt.title('Order Frequency by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Orders')
plt.xticks(rotation=0)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
